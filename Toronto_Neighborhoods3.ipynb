Import Pandas library

```python
import pandas as pd
```

Create Toronto dataframe from csv file

```python
Toronto = pd.read_csv('Toronto-Neighborhoods.csv')
```

Show dataframe 10 first records

```python
Toronto.head(10)
```

Remove Boroughs with name "Not assigned"

```python
Notassigned=Toronto[Toronto['Borough']=='Not assigned'].index
Toronto.drop(Notassigned, axis=0, inplace=True)
Toronto.head(10)
```

Shape of dataframe

```python
Toronto.shape
```

Import geographical coordinates of Toronto

```python
Coordinates = pd.read_csv('Geospatial_Coordinates.csv')
Coordinates.head()
```

Merge Toronto and Coordinates dataframes in a new dataframe

```python
Toronto2 = pd.merge(Toronto, Coordinates, on ='Postal Code')
Toronto2.head(10)
```

```python
Toronto2.shape
```

Number of boroughs and neighborhoods of Toronto

```python
print('Toronto2 dataframe has {} boroughs and {} neighborhoods.'.format(
        len(Toronto2['Borough'].unique()),
        Toronto2.shape[0]
    )
)
```

Install and import libraries to graph maps

```python
!pip install geopy

from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

import folium # map rendering library
```

Set geograpical coordinates of Toronto

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="Tor_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(Toronto2['Latitude'], Toronto2['Longitude'], Toronto2['Borough'], Toronto2['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='red',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
